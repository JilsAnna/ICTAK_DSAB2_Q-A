# Read Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

data=pd.read_excel("E:\ICTAK-DSA\Telco-Customer-Churn.xlsx")
df=pd.DataFrame(data)
df
```

# Compare churn count with respect to gender

```python
df['gender'].count()
plt.figure(figsize=(8,8))
sns.countplot(x='Churn',data=df,hue='gender')
plt.title('Comparison Of Churn Count Respect To  Gender',fontsize=15)
```

# Find out how many female senior citizens there in the dataset

```python
fsc=df[(df['SeniorCitizen']==1) & (df['gender']=='Female')]
fsc1=sns.countplot(fsc['gender'])
plt.title('Details of Female Senior Citizens ',fontsize=14)
for p in fsc1.patches:
    fsc1.text(p.get_x()+.10,p.get_height()-40,str(p.get_height()),fontsize=12)
```

# Compare 'tenure' with 'Total Charges'

```python
sns.scatterplot(x=df['tenure'],y=df['TotalCharges'])
plt.title('Tenure with Total Charges',fontsize=10)
plt.xlabel('Tenure',fontsize=12)
plt.ylabel('Total Charges',fontsize=12)
```

# Find out which contract is preferred by the senior citizen

```python
csc = df[df["SeniorCitizen"]==1]
plt.figure(figsize=(10,10))
plt.title(" Distribution of Contract is  preferred by the Senior Citizens ", fontsize=14)
csc.Contract.value_counts(sort=True).plot.pie(fontsize=14,startangle = 90,autopct = '%1.1f')
plt.show()
```

# Findings  on Payment Method

```python
plt.figure(figsize=(12,12))
sns.countplot(df['PaymentMethod'],hue=df['gender'])
plt.xlabel('Payment Method',fontsize=14)
plt.ylabel('Count',fontsize=14)
plt.yticks(fontsize=10)
plt.xticks(fontsize=10)
```

Findings: The Most Preferred way of payment among the customers is ELectronic Check.BAnk Transfer is least used by male category           and credit card for female category.

```python
payment = df["PaymentMethod"].value_counts()
plt.figure(figsize=(10,10))
payment.plot.pie(explode=(0.1,0.1,0.1,0.1),autopct='%1.2f%%')
plt.title("Payment Method Distribution Details in Percentage ",fontsize = 12)
plt.ylabel("")
plt.show()
```

Findings: This pie plot helps to understand the percentage distribution of  various payment methods among cutomers in car data
