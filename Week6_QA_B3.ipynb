```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("E:\ICTAK-DSA\Dataset_w6.xlsx")
df
```

# Find out how many countries are there in the dataset.


```python

print("Number countries are there in the dataset:",df.country.count())
```

# Examine the data by using visualizations.

```python
plt.figure(figsize=(16,10))
sns.countplot(x='Region',data=df)
plt.title('Count of COuntries in a Region')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(30,22))
plt.title("Top 5 Countries w.r.t Population in 2017",fontsize=26)
sns.barplot(x='country', y='Population in thousands (2017)', data=df.nlargest(5,'Population in thousands (2017)'))
```

```python
plt.figure(figsize=(14,10))
plt.title("Distribution of GDP in various Region")
sns.barplot(x='Region', y='GDP per capita (current US$)', data = df)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(14,10))
plt.title("Distribution of International Trade Imports in various Region")
sns.boxplot(x='Region', y='International trade: Imports (million US$)', data = df)
plt.xticks(rotation=90)
```

# Find out how many countries with respect to the region are available in the dataset

```python
df['Region'].value_counts()
```

# On average which region has the highest
a. GDP per capita (current US$)

```python
high_gdp=df.groupby('Region')['GDP per capita (current US$)'].mean()
high_gdp.nlargest(1)
```

b. International trade: Imports (million US$)

```python
high_imports=df.groupby('Region')['International trade: Imports (million US$)'].mean()
high_imports.nlargest(1)
```

# Find out which region has more consistent with respect to GDP percapita (current US$)

```python
std1 = df.groupby('Region')['GDP per capita (current US$)'].std() 
print("Region has more consistent with respect to GDP percapita (current US$) is:")
std1.nsmallest(1)
```
